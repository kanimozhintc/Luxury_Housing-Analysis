```python
#Convert Ticket_price column to float value:
df['Ticket_Price_Cr'] =(df['Ticket_Price_Cr'].astype(str).replace(r'[₹,]|Cr|Lakh|M', '', regex=True).str.strip().astype(float))
```

```python
#convert Purchase quarter column to datetime format:
import numpy as np
df['Purchase_Quarter']=pd.to_datetime(df['Purchase_Quarter'],errors='coerce')
df['NRI_Buyer'] = df['NRI_Buyer'].map({'Yes': True, 'No': False})
df['NRI_Buyer'].fillna(False, inplace=True)
```

```python
#filling null values for Ticket_Price_Cr:
df['Ticket_Price_Cr'].fillna(df["Ticket_Price_Cr"].median(),inplace=True)
#filling null values for Buyer_Comments  :
df['Buyer_Comments'].fillna("unknown",inplace=True)
#filling null values for Amenity_Score:
df['Amenity_Score'].fillna(df["Amenity_Score"].median(),inplace=True)
```

```python
df['Developer_Name'] = df['Developer_Name'].str.strip().str.lower()
df['Micro_Market'] = df['Micro_Market'].str.strip().str.title()

df['Quarter_Number'] = pd.to_datetime(df['Purchase_Quarter']).dt.quarter

```

```python
df['Ticket_Price_Cr'] = (
    df['Ticket_Price_Cr']
    .astype(str)                         # make sure it's string for cleaning
    .str.replace("Cr", "", regex=False)  # remove "Cr"
    .str.replace(",", "", regex=False)   # remove commas
    .str.strip()                         # remove spaces
)
df['Ticket_Price_Cr'] = pd.to_numeric(df['Ticket_Price_Cr'], errors="coerce")
df = df[df['Unit_Size_Sqft'] > 0]  # avoid division by zero
df['Price_per_Sqft'] = (df['Ticket_Price_Cr'] * 1e7) / df['Unit_Size_Sqft']
```
